```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'ecommerce.csv')
df
```

```python
# 1- What are the total Order value for each country
df.groupby('Country')['OrderValue'].sum().reset_index()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['Year-Month'])
```

```python
# 2- What each Major Category sold quantity during November 2011
nov =df[df['date'] == '2011-11']
nov.groupby('Major Category')['Quantit'].sum().reset_index()
```

```python
df
```

```python
df['month']=df.date.dt.month
```

```python
# 4- What Month/s contributes to high demand in orders
df.groupby('month')['Quantit'].sum().nlargest()
```

```python
# 5- What Month/s contributes to high OrderValue?
df.groupby('month')['OrderValue'].sum().nlargest()
```

```python
# 6- Why April and December suffer from low sales? 
low_mon = df[df['month'].isin([4, 12])]
low_mon.groupby('month').describe().T
```

```python
# 7- Why May and November have high sales? 
high_mon = df[df['month'].isin([5, 11])]
high_mon.groupby('month').describe().T
```

```python
# 8- Most Major Items sold in April and December
low_mon.groupby('Minor Category')['Quantit'].sum().sort_values(ascending=0).reset_index()
```

```python
# 9- Most Major Items sold in May and November
high_mon.groupby('Minor Category')['Quantit'].sum().sort_values(ascending=0).reset_index()
```

```python
# 10- What Hour recives high demand for ordering?
df['Hour'] = df['InvoiceTime'].str.split(':').str[0]
df.groupby('Hour')['Quantit'].sum().sort_values(ascending=0).reset_index()
```

```python
# 11- What's the Time of the day that contributes to high Order Values?
def time_of_day(x):
    if x in range(6,12):
        return 'Morning'
    elif x in range(12,16):
        return 'Afternoon'
    elif x in range(16,22):
        return 'Evening'
    else:
        return 'Late night'
df.Hour=df.Hour.astype('int')    
df['OrderTimeOfDay']=df['Hour'].apply(time_of_day)
df.groupby('OrderTimeOfDay')['Quantit'].sum().sort_values(ascending=0).reset_index()
```

```python
# 12 - Order VAlue For each Country in each month
pd.pivot_table(df,index='Country',columns='month',values='OrderValue')
```

```python
# 13 - Order VAlue For each Country and Major Category in Minor Category
pd.pivot_table(df,index=['Major Category','Minor Category'],columns='Country',values='OrderValue',aggfunc=np.sum)
```

```python
# 15- Correlation between 'Quantit','OrderValue','UnitPrice'.
features=['Quantit','OrderValue','UnitPrice']
df[features].corr()
```
