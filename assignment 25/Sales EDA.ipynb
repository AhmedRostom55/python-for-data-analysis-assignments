```python
for v in df['Quantity Ordered'].unique():
    print(v)
```

```python
def fix_qty(qty_str):
    try:
        return int(qty_str)
    except:
        return np.nan
        

df['Quantity Ordered'] = df['Quantity Ordered'].apply(fix_qty)
df.info()
```

```python
df
```

```python
for v in df['Price Each'].unique():
    print(v)
```

```python
def fix_price_each(price_str):
    try:
        return float(price_str[:-1])
    except:
        return np.nan
    
df['Price Each'] = df['Price Each'].apply(fix_price_each)
df
```

# Feature Engineering

### Profit
Profit = cash - taxes (0.14)

```python
def get_profit(row):
    qty = row['Quantity Ordered']
    price = row['Price Each']
    cash = qty * price
    taxes = cash * 0.14
    profit = cash - taxes
    return profit
    


df['Profit'] = df.apply(get_profit, axis=1)
df
```

```python
def get_city(address):
    try:
        return address.split(', ')[1]
    except:
        return np.nan
    

# df['City'] = df['Purchase Address'].apply(lambda address: address.split(', ')[1])

df['City'] = df['Purchase Address'].apply(get_city)
df
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M', errors='coerce')
df.info()
```

```python
df
```

```python
df['Weekday'] = df['Order Date'].dt.day_name()
df
```
