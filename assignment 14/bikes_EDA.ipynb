## Profit for each day

```python
df_eda.head()
```

```python
df_eda['day']= df_eda['datetime'].dt.day
df_eda['day_of_week']= df_eda['datetime'].dt.day_name()
df_eda['month']= df_eda['datetime'].dt.month
df_eda['month_name']= df_eda['datetime'].dt.month_name()
df_eda['year']= df_eda['datetime'].dt.year
df_eda['hour']= df_eda['datetime'].dt.hour
```

```python
df_eda.head()
```

```python
df_eda.groupby('day')['Profit'].sum().sort_values(ascending=0).round().reset_index()
```

```python
df_eda.groupby('day_of_week')['Profit'].sum().sort_values(ascending=0).round().reset_index()
```

## Which season that has more registration & profit.

## Which weather condition that has more registration & profit.

```python
df_eda.groupby('weather')[['Profit','registered']].sum().reset_index()
```
